## Popularity Based Recommendation System
In this recommendation system it will select the books with highest ratings, like the books that has more than 250 ratings.

To do this we need to merge the ratings and the books tables on the ISBN column to find link the book name and the average ratings for that book.

```python
ratings_bookname = ratings.merge(books, on='ISBN')
ratings_bookname
```

### Number of Ratings for each Book

```python
ratings_book_count = ratings_bookname.groupby('Book-Title').count()['Book-Rating'].reset_index()
ratings_book_count = ratings_book_count.rename(columns = {'Book-Rating': 'Number-of-Ratings'})
ratings_book_count
```

### Average Rating for each Book

```python
ratings_book_avg = ratings_bookname.groupby('Book-Title').mean()['Book-Rating'].reset_index()
ratings_book_avg = ratings_book_avg.rename(columns = {'Book-Rating': 'Average-Rating'})
ratings_book_avg
```

```python
popularity_books = ratings_book_count.merge(ratings_book_avg, on = 'Book-Title')
popularity_books
```

```python
popular_books = popularity_books[popularity_books['Number-of-Ratings']>250].sort_values('Average-Rating', ascending = False)
popular_books
```

```python
popular_books.merge(books, on = 'Book-Title').drop_duplicates('Book-Title').shape
```

```python
popular_final = popular_books.merge(books, on = 'Book-Title').drop_duplicates('Book-Title')[['Book-Title', 'Book-Author', 'Image-URL-M', 'Number-of-Ratings', 'Average-Rating']]
popular_final
```

## Collaborative Filtering Recommendation System

Here, the users and the books will be tabulated as columns and rows respectively. By doing that we can find the similarities between different pairs of users or books which will help us in making proper recommendations to the users.

For proper recommendations, we can use a condition that the users considered here should be some experienced readers, meaning they should have at least rated a 100 books. And for the books they should have a minimum number of ratings of 50.

```python
ratings_count = (ratings_bookname.groupby('User-ID').count()['Book-Rating'] > 200)
user_index = ratings_count[ratings_count].index
user_index
```

```python
users_filtered = ratings_bookname[ratings_bookname['User-ID'].isin(user_index)]
```

```python
books_count = users_filtered.groupby('Book-Title').count()['Book-Rating'] >= 50
books_index = books_count[books_count].index
books_index
```

```python
final_ratings = users_filtered[users_filtered['Book-Title'].isin(books_index)]
final_ratings
```

```python
user_books = final_ratings.pivot_table(index = 'Book-Title', columns = 'User-ID', values = 'Book-Rating')
user_books.fillna(0, inplace=True)
user_books
```

### Cosine Similarity Scores
The Cosine similarity scores are used to identify which books are similar to each other.

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim_scores = cosine_similarity(user_books)
sim_scores
```

### Recommend the top 8 books
A function can be defined that can recommend the top 8 books when the user enters a book name.

```python
def recommend_books(book_name):
    index = np.where(user_books.index == book_name)[0][0]
    similar_books = sorted(list(enumerate(sim_scores[index])), key=lambda x:x[1], reverse=True)[1:9]
    
    for i in similar_books:
        
        print(user_books.index[i[0]])
```

```python
recommend_books('Message in a Bottle')
```
